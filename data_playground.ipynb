```python
import os

import numpy as np
import matplotlib.pyplot as plt

import nibabel as nib
```

```python
files_path = "data/US_images/3_Spine/"
```

```python
leg_img = nib.load("data/US_images/1_Leg_LAV_SA.nii.gz")
leg_img = leg_img.get_fdata()

print(leg_img.shape)
```

```python
leg_label = nib.load("data/Labels/1_Leg_LAV_SA_labels.nii")
leg_label = leg_label.get_fdata()

print(leg_label.shape)
```

```python
random_idxs = np.random.randint(0, leg_img.shape[2], 5)

for plt_idx, idx in enumerate(random_idxs):
    plt.subplot(1, 5, plt_idx + 1)
    plt.imshow(leg_img[:, :, idx], cmap="gray")
plt.show()

for plt_idx, idx in enumerate(random_idxs):
    plt.subplot(1, 5, plt_idx + 1)
    plt.imshow(leg_label[:, :, 0, idx].T, cmap="gray")
plt.show()
```
