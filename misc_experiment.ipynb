```python
import matplotlib.pyplot as plt
```

```python
xs = [10, 100]
ys = [9.787, 42.654]
```

```python
plt.scatter(xs, ys)
plt.show()
```

```python
# Fit linear model
import numpy as np

X = np.array(xs).reshape(-1, 1)
y = np.array(ys)

from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X, y)

print(model.coef_, model.intercept_)
```

```python
model.predict([[1714]])
```
