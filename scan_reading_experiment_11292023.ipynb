```python
import numpy as np
import nibabel as nib

import matplotlib.pyplot as plt
```

```python
label_path = "data/whole_body_labeled.nii"
```

```python
labels = nib.load(label_path).get_fdata()
```

```python
# Visualize each label as point cloud

for label_idx in np.unique(labels):
    if label_idx == 0:
        continue
    label = labels == label_idx
    label = label.astype(np.float32)
    label[label == 0] = np.nan

    plt.imshow(label[:, :, 100])

    plt.title(f"Label {label_idx}")
    plt.show()
```
